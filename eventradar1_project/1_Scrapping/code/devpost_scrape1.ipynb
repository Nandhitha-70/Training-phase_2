```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import pandas as pd
import time
CHROMEDRIVER_PATH = r"C:\Users\nandhu S\OneDrive\Desktop\eventradar_project\chromedriver.exe"
options = Options()
options.add_argument("--headless")  
options.add_argument("--disable-gpu")

service = Service(CHROMEDRIVER_PATH)
driver = webdriver.Chrome(service=service, options=options)
```

```python
base_url = "https://devpost.com/hackathons?status[]=upcoming&page="
all_data = []

for page_num in range(1, 11): 
    driver.get(base_url + str(page_num))
    time.sleep(5)

    soup = BeautifulSoup(driver.page_source, "html.parser")
    cards = soup.find_all("a", class_="tile-anchor")

    for card in cards:
        try:
            title = card.find("h3").text.strip()
            url = card["href"]
            if not url.startswith("http"):
                url = "https://devpost.com" + url

            platform = "Devpost"
            event_type = "Hackathon"

            location_elem = card.find("div", class_="info")
            location = location_elem.text.strip() if location_elem else "Unknown"

            date_elem = card.find("div", class_="submission-period")
            event_date = date_elem.text.strip() if date_elem else "TBD"

            tags = [tag.get("title") for tag in card.select(".theme-label")]
            topic_tags = ", ".join(tags) if tags else "Uncategorized"

            host_elem = card.find("span", class_="host-label")
            speakers = host_elem.get("title") if host_elem else ""

            all_data.append({
                "event_title": title,
                "platform": platform,
                "topic_tags": topic_tags,
                "event_type": event_type,
                "event_date": event_date,
                "location": location,
                "speakers_orgs": speakers,
                "url": url,
                "audience_segment": "" 
            })

        except Exception as e:
            print(f"‚ùå Skipping card due to: {e}")
```

```python
df = pd.DataFrame(all_data)
df.to_csv("step1_raw_devpost_events1.csv", index=False)
print(f"‚úÖ Scraped {len(df)} events and saved to CSV.")
df.head()
```
