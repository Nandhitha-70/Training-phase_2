```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
import time
chrome_driver_path = "C:\\Users\\nandhu S\\OneDrive\\Desktop\\eventradar_project\\chromedriver.exe"
options = webdriver.ChromeOptions()
options.add_argument("--headless")
service = Service(chrome_driver_path)
driver = webdriver.Chrome(service=service, options=options)
url = "https://lu.ma/bengaluru"
driver.get(url)
time.sleep(5)

soup = BeautifulSoup(driver.page_source, "lxml")
cards = soup.select("a.event-link.content-link")
event_list = []

for card in cards:
    relative_link = card.get("href", "")
    full_link = "https://lu.ma" + relative_link
    driver.get(full_link)
    time.sleep(3)
    event_soup = BeautifulSoup(driver.page_source, "lxml")
    title = event_soup.select_one("h1").text.strip() if event_soup.select_one("h1") else ""
    
    date_block = event_soup.find("div", class_="jsx-3191908726 date")
    event_date = date_block.text.strip() if date_block else ""
    location_block = event_soup.find("div", string=lambda x: x and "Location" in x)
    location = location_block.find_next("div").text.strip() if location_block else "Online"
    desc = event_soup.select_one("div.markdown").text.strip() if event_soup.select_one("div.markdown") else ""

    event_list.append({
        "event_title": title,
        "platform": "Luma",
        "topic_tags": "",      
        "event_type": "Meetup", 
        "event_date": event_date,
        "location": location,
        "speakers_orgs": "",    
        "url": full_link,
        "audience_segment": ""  
    })

driver.quit()
luma_raw_df = pd.DataFrame(event_list)
luma_raw_df.to_csv("01_raw_luma_events.csv", index=False)
luma_raw_df.head()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
import time
chrome_path = r"C:\Users\nandhu S\OneDrive\Desktop\eventradar_project\chromedriver.exe"
service = Service(chrome_path)
driver = webdriver.Chrome(service=service)

url = "https://lu.ma/singapore"
driver.get(url)
time.sleep(3) 
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(driver.page_source, "lxml")

event_links = soup.find_all("a", class_="event-link content-link")
print(f"Found {len(event_links)} events")

base_url = "https://lu.ma"
events = []

for link in event_links:
    try:
        title = link.get("aria-label", "").strip()
        href = base_url + link.get("href", "").strip()
        
        events.append({
            "event_title": title,
            "platform": "Luma",
            "event_type": "",  
            "event_date": "",  
            "location": "",    
            "url": href,
            "speakers_orgs": "", 
            "topic_tags": "",     
            "audience_segment": ""  
        })
    except Exception as e:
        print(f"Error: {e}")

import pandas as pd
df_raw = pd.DataFrame(events)
df_raw.to_csv("eventradar_raw_luma.csv", index=False)
df_raw.head()
```
