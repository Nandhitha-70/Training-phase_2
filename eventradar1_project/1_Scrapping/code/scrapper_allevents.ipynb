```python
# üì¶ Required Libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
import pandas as pd
import time

# üìç Set up ChromeDriver
service = Service(executable_path=chrome_path)
driver = webdriver.Chrome(service=service)

# üåê Load the target URL
url = "https://allevents.in/coimbatore/business"
driver.get(url)
time.sleep(5)  # wait for content to load

# üç≤ Parse with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')
event_cards = soup.find_all("li", class_="event-card")

# üßπ Extract required fields
data = []
for card in event_cards:
    try:
        title = card['data-name'].strip()
        event_url = card['data-link'].strip()
        location = card.find("div", class_="subtitle").text.strip()
        date = card.find("div", class_="date").text.strip()
        price_tag = card.find("span", class_="price")
        price = price_tag.text.strip() if price_tag else "Free"

        data.append({
            "event_title": title,
            "platform": "AllEvents",
            "event_date_raw": date,
            "location": location,
            "url": event_url,
            "price": price
        })
    except Exception as e:
        print("Skipping a card due to error:", e)

# ‚úÖ Close driver
driver.quit()

# üìÅ Save to CSV
df = pd.DataFrame(data)
df.to_csv("1_raw_allevents_data.csv", index=False)
df.head()
```

```python
# üìö Imports
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
import time

# üõ†Ô∏è Set up ChromeDriver
from selenium.webdriver.chrome.service import Service
chrome_path = r"C:\Users\nandhu S\OneDrive\Desktop\eventradar_project\chromedriver.exe"
service = Service(executable_path=chrome_path)
driver = webdriver.Chrome(service=service)

# üåê Open AllEvents Bangalore Business Page
url = "https://allevents.in/bangalore/business?ref=eventlist-category-filter#search"
driver.get(url)
time.sleep(5)  # allow JS to load

# üçú Parse HTML with BeautifulSoup
soup = BeautifulSoup(driver.page_source, "html.parser")
event_cards = soup.find_all("li", class_="event-card")

# üì¶ Extract Event Data
events = []

for card in event_cards:
    try:
        event_title = card.find("h3").text.strip()
        event_url = card.get("data-link", "").strip()
        location = card.find("div", class_="subtitle").text.strip()
        date = card.find("div", class_="date").text.strip()
        price_tag = card.find("span", class_="price")
        price = price_tag.text.strip() if price_tag else "Free"

        events.append({
            "event_title": event_title,
            "platform": "AllEvents",
            "event_date_raw": date,
            "location": location,
            "url": event_url,
            "price": price
        })
    except Exception as e:
        print(f"Skipping event due to error: {e}")

# ‚úÖ Close the driver
driver.quit()

# üíæ Save raw data
df_bangalore = pd.DataFrame(events)
df_bangalore.to_csv("1_raw_allevents_bangalore_data.csv", index=False)
df_bangalore.head()
```
