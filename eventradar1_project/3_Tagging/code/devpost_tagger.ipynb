```python
import pandas as pd

# ðŸ“¥ Load cleaned dataset
df = pd.read_csv("step2_cleaned_devpost_events.csv")
```

```python
# Define simple keyword-topic mapping
topic_keywords = {
    "AI": ["ai", "machine learning", "ml"],
    "SaaS": ["saas"],
    "Design": ["design", "ux", "ui"],
    "EdTech": ["education", "student", "learning"],
    "Fintech": ["fintech", "finance", "bank", "crypto"],
    "NoCode": ["no code", "low code"],
    "Web3": ["web3", "blockchain"],
    "Open Source": ["open source", "github"],
    "Productivity": ["productivity", "efficiency", "tools"]
}

def classify_topics(text):
    text = str(text).lower()
    topics = []
    for topic, keywords in topic_keywords.items():
        if any(word in text for word in keywords):
            topics.append(topic)
    return ", ".join(topics) if topics else "General Tech"

df["topic_tags"] = df["event_title"].apply(classify_topics)
```

```python
def infer_audience(title, topics):
    title = title.lower()
    topics = topics.lower()

    if "student" in title or "education" in topics:
        return "Students"
    elif "developer" in title or "code" in topics or "open source" in topics:
        return "Developers"
    elif "design" in topics:
        return "Designers"
    elif "marketing" in title or "productivity" in topics:
        return "Marketers"
    elif "startup" in title or "founder" in title:
        return "Founders"
    else:
        return "General Tech"

df["audience_segment"] = df.apply(lambda row: infer_audience(row["event_title"], row["topic_tags"]), axis=1)
```

```python
df.to_csv("step3_tagged_devpost_events.csv", index=False)
df.head()
```
