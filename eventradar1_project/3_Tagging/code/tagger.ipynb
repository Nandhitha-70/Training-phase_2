```python
# ‚úÖ Step 3: Add Tags Based on Title and URL
import pandas as pd

# Load cleaned data
df = pd.read_csv("02_cleaned_luma_events.csv")

# üîñ Tag topics based on keywords in title
def tag_topics(title):
    title = title.lower()
    tags = []

    if "ai" in title or "ml" in title or "machine learning" in title:
        tags.append("AI")
    if "design" in title or "ux" in title:
        tags.append("UX")
    if "gtm" in title or "growth" in title:
        tags.append("GTM")
    if "startup" in title or "pitch" in title or "founder" in title:
        tags.append("Startup")
    if "saas" in title:
        tags.append("SaaS")
    if "open source" in title:
        tags.append("Open Source")
    if "fintech" in title:
        tags.append("Fintech")
    if "nocode" in title or "lowcode" in title:
        tags.append("NoCode")

    return ", ".join(tags) if tags else "General Tech"

# üë• Tag audience segments
def tag_audience(title):
    title = title.lower()
    if "developer" in title or "engineer" in title:
        return "Developers"
    elif "designer" in title or "ux" in title:
        return "Designers"
    elif "marketer" in title or "growth" in title:
        return "Marketers"
    elif "founder" in title or "startup" in title:
        return "Founders"
    elif "student" in title or "college" in title:
        return "Students"
    return "Tech Enthusiasts"

df["topic_tags"] = df["event_title"].apply(tag_topics)
df["audience_segment"] = df["event_title"].apply(tag_audience)

# Save tagged output
df.to_csv("03_tagged_luma_events.csv", index=False)
df.head()
```
