```python
# üìö Import libraries
import pandas as pd
from datetime import datetime
import re

# üì• Load raw CSV
df = pd.read_csv("1_raw_allevents_data.csv")

# üßπ Clean and format event_date
def parse_date(text):
    try:
        match = re.search(r"(\d{1,2})\s([A-Za-z]+)", text)
        if match:
            day = match.group(1)
            month = match.group(2)
            year = "2025"
            return datetime.strptime(f"{day} {month} {year}", "%d %b %Y").strftime("%d-%m-%Y")
    except:
        return None

df["event_date"] = df["event_date_raw"].apply(parse_date)

# üè∑Ô∏è Standard fields
df["event_type"] = "Networking"
df["speakers_orgs"] = ""
df["topic_tags"] = ""
df["audience_segment"] = ""

# ‚úÇÔ∏è Drop raw date
df_final = df[[
    "event_title", "platform", "topic_tags", "event_type", "event_date",
    "location", "speakers_orgs", "url", "audience_segment"
]]

# üíæ Save cleaned data
df_final.to_csv("2_cleaned_allevents_data.csv", index=False)
df_final.head()
```
