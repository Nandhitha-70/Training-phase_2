```python
# ðŸ“š Load dependencies
import pandas as pd

# ðŸ“¥ Load cleaned data
df = pd.read_csv("2_cleaned_allevents_data.csv")

# ðŸ§  Infer topic tags
def infer_topic(text):
    topics = {
        "Startup": ["startup", "entrepreneur", "pitch"],
        "Funding": ["investor", "vc", "funding"],
        "Networking": ["network", "mixer", "connect"],
        "Tech": ["tech", "software", "ai", "cloud", "developer"],
        "Marketing": ["marketing", "seo", "growth"],
        "Business": ["business", "scale", "growth", "strategy"]
    }
    for topic, keywords in topics.items():
        if any(kw.lower() in text.lower() for kw in keywords):
            return topic
    return "General"

df["topic_tags"] = df["event_title"].apply(infer_topic)

# ðŸŽ¯ Infer audience segment
def infer_audience(text):
    if "developer" in text.lower() or "tech" in text.lower():
        return "Developers"
    if "student" in text.lower():
        return "Students"
    if "startup" in text.lower() or "founder" in text.lower() or "entrepreneur" in text.lower():
        return "Founders"
    if "marketing" in text.lower():
        return "Marketers"
    return "Business Audience"

df["audience_segment"] = df["event_title"].apply(infer_audience)

# ðŸ’¾ Save tagged data
df.to_csv("3_tagged_allevents_data.csv", index=False)
df.head()
```

```python
# ðŸ“š Import dependencies
import pandas as pd

# ðŸ“¥ Load cleaned Bangalore data
df = pd.read_csv("2_cleaned_allevents_bangalore_data.csv")

# ðŸ§  Infer topic tags
def infer_topic(text):
    topics = {
        "AI": ["ai", "artificial intelligence"],
        "Innovation": ["summit", "conference", "expo", "future"],
        "Startup": ["startup", "entrepreneur"],
        "Tech": ["tech", "software", "cloud"],
        "Business": ["business", "scale", "strategy", "network"],
    }
    for tag, keywords in topics.items():
        if any(word in text.lower() for word in keywords):
            return tag
    return "General"

df["topic_tags"] = df["event_title"].apply(infer_topic)

# ðŸŽ¯ Infer audience segment
def infer_audience(text):
    if "student" in text.lower():
        return "Students"
    if "developer" in text.lower() or "tech" in text.lower():
        return "Developers"
    if "startup" in text.lower() or "founder" in text.lower() or "entrepreneur" in text.lower():
        return "Founders"
    if "marketing" in text.lower():
        return "Marketers"
    return "Business Audience"

df["audience_segment"] = df["event_title"].apply(infer_audience)

# ðŸ’¾ Save tagged Bangalore event data
df.to_csv("3_tagged_allevents_bangalore_data.csv", index=False)
df.head()
```
