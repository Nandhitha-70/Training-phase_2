```python
#  Step 4: Visualize Tagged Luma Events
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the tagged dataset
df = pd.read_csv("03_tagged_luma_events.csv")

# Set style
sns.set(style="whitegrid")

# 1 Bar chart: Number of events per Topic
plt.figure(figsize=(10, 5))
sns.countplot(data=df, y="topic_tags", order=df["topic_tags"].value_counts().index)
plt.title("Events by Topic (Luma Bengaluru)")
plt.xlabel("Number of Events")
plt.ylabel("Topic")
plt.tight_layout()
plt.savefig("luma_topic_distribution.png")
plt.show()

# 2 Pie chart: Online vs In-person
plt.figure(figsize=(6, 6))
df["location"].value_counts().plot.pie(autopct='%1.1f%%', startangle=140)
plt.title("Event Format Distribution")
plt.ylabel("")
plt.tight_layout()
plt.savefig("luma_format_distribution.png")
plt.show()

# 3 Bar chart: Audience Segment
plt.figure(figsize=(10, 4))
sns.countplot(data=df, x="audience_segment", order=df["audience_segment"].value_counts().index)
plt.title("Audience Segment Distribution")
plt.xlabel("Audience")
plt.ylabel("Number of Events")
plt.tight_layout()
plt.savefig("luma_audience_segment.png")
plt.show()
```
